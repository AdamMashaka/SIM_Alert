```python
import os
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Define the path to the training data directory
train_data = 'extraction/train_data'

# Create the directory if it doesn't exist
os.makedirs(train_data, exist_ok=True)

# Function to generate random timestamps
def generate_timestamps(start, end, n):
    start_u = start.timestamp()
    end_u = end.timestamp()
    return [datetime.fromtimestamp(ts) for ts in np.random.uniform(start_u, end_u, n)]

# Generate sample data for access_logs.csv
access_logs_data = {
    'timestamp': generate_timestamps(datetime(2023, 1, 1), datetime(2023, 12, 31), 1000),
    'user_id': np.random.randint(1, 100, 1000),
    'action': np.random.choice(['login', 'access', 'logout'], 1000),
    'resource': np.random.choice(['server1', 'server2', 'file1', 'file2'], 1000),
    'label': np.random.choice(['authorized', 'unauthorized'], 1000)
}
access_logs_df = pd.DataFrame(access_logs_data)
access_logs_df.to_csv(os.path.join(train_data, 'access_logs.csv'), index=False)

# Generate sample data for auth_logs.csv
auth_logs_data = {
    'timestamp': generate_timestamps(datetime(2023, 1, 1), datetime(2023, 12, 31), 1000),
    'user_id': np.random.randint(1, 100, 1000),
    'auth_method': np.random.choice(['password', 'fingerprint', 'face_recognition'], 1000),
    'status': np.random.choice(['success', 'failure'], 1000)
}
auth_logs_df = pd.DataFrame(auth_logs_data)
auth_logs_df.to_csv(os.path.join(train_data, 'auth_logs.csv'), index=False)

# Generate sample data for network_logs.csv
network_logs_data = {
    'timestamp': generate_timestamps(datetime(2023, 1, 1), datetime(2023, 12, 31), 1000),
    'user_id': np.random.randint(1, 100, 1000),
    'source_ip': np.random.choice(['192.168.1.1', '192.168.1.2', '10.0.0.1', '10.0.0.2'], 1000),
    'destination_ip': np.random.choice(['192.168.1.3', '192.168.1.4', '10.0.0.3', '10.0.0.4'], 1000),
    'action': np.random.choice(['allow', 'deny'], 1000)
}
network_logs_df = pd.DataFrame(network_logs_data)
network_logs_df.to_csv(os.path.join(train_data, 'network_logs.csv'), index=False)

print("Sample data created for access_logs.csv, auth_logs.csv, and network_logs.csv")
```

```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

def load_and_preprocess_images(image_dir):
    images = []
    labels = []
    for filename in os.listdir(image_dir):
        if filename.endswith('.png') or filename.endswith('.jpg'):
            img = cv2.imread(os.path.join(image_dir, filename), cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (128, 128))
            img = img / 255.0
            images.append(img)
            labels.append(filename.split('_')[0])  # Assuming filename format: 'label_xxx.png'
    return np.array(images), np.array(labels)

image_dir = '/path/to/fingerprint/images'  # Replace with the actual path
images, labels = load_and_preprocess_images(image_dir)
images = images.reshape(-1, 128, 128, 1)  # Reshape for CNN input

# Encode labels
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(labels)
labels = to_categorical(labels)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)
```
